# Types of Variables:

```python
1. Global variable
2. Local Variable
```

```python
1. Global Variable: 
    If we declare variable outside of the function is called global variable.
```

```python
a=10 #global variable
def fun1():
    print("a value from function1:",a)
def fun2():
    print("a value from function2:",a)
```

```python
fun1()
```

```python
fun2()
```

2. Local Variable

if we declare or initialize the variable within the function is called local variable

```python
def fun1():
    b=10 #local variable
    print("a value from function1:",b)
def fun2():
    print("a value from function2:",b)
```

```python
fun1()
```

```python
fun2()
```

# Factorial of given number

```python
#program to find the factorial of given number
def fact(n):
    if n==0:
        result=1
    else:
        result=n*fact(n-1)
    return result
```

```python
fact(4)
```

```python
fact(3)
```

# Anonymous Function

```python
Function without name is called anonymous function

The main use of anonymous function is for one time usage
```

```python
#a normal function creation
def Square(n):
    return n*n
Square(2)
```

```python
lambda is used to create anonymous functions
```

```python
lambda variables_list: Expression
```
